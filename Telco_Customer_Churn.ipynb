```python
#comparing churn by senior citizen
#count the total no of senior citizen(male,female) . For that i am using count plot.
plt.figure(figsize=(4,4),facecolor="lightblue")
plt.title("Count of SeniorCitizen",color="brown",fontweight="bold")
a=sns.countplot(x=df["SeniorCitizen"],data=df)
a.bar_label(a.containers[0])
plt.xlabel("SeniorCitizen",color="brown",fontweight="bold",fontsize=12)
plt.xticks(color="black",fontweight="bold",fontsize=8)
plt.ylabel("Count",color="brown",fontweight="bold",fontsize=12)
plt.yticks(color="black",fontweight="bold",fontsize=8)
plt.gca().set_facecolor("beige")
plt.show()
```

```python

ct = pd.crosstab(df["SeniorCitizen"], df["Churn"])

# Convert to percentage
ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
ax = ct_percent.plot(kind="bar", stacked=True, figsize=(4,4),
                     color=["lightblue", "#fc8d62"])
#66c2a5

# Title and labels
plt.title("Churn by SeniorCitizen (%)", color="brown", fontweight="bold")
plt.xlabel("SeniorCitizen", color="brown", fontweight="bold")
plt.ylabel("Percentage", color="brown", fontweight="bold")

# Add percentage labels
for i in range(len(ct_percent)):
    cumulative = 0
    for j in range(len(ct_percent.columns)):
        value = ct_percent.iloc[i, j]
        if value > 0:
            ax.text(i, cumulative + value/2, f"{value:.1f}%",
                    ha='center', va='center', fontsize=8, fontweight="bold")
            cumulative += value

plt.xticks(rotation=0)
plt.legend(title="Churn")
plt.gca().set_facecolor("beige")
plt.tight_layout()
plt.show()
```

From the above comparision we can get to know that a greater percenatge of people in senior citizen category have curned.
